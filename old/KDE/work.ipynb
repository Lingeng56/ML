```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_kde(x, data, bandwidth=0.1):
    import math
    gauss = lambda x_: (1/math.sqrt(2*math.pi))*math.exp(-0.5*(x_**2))
    n = len(data)
    res = 0
    if n == 0:
        return 0
    for i in range(len(data)):
        res += gauss((x-data[i])/bandwidth)
    res /= (n*bandwidth)

    return res
```

```python
inputs = np.random.randn(20000).tolist()
bandwidth_ = 1.05*np.std(inputs)*(len(inputs)**(-1/5))
x_array = np.linspace(min(inputs), max(inputs), 50)
y_array = [get_kde(x_array[i], inputs, bandwidth_) for i in range(x_array.shape[0])]

plt.figure()
plt.hist(inputs, bins=40, density=True)
plt.plot(x_array.tolist(), y_array, color='red', linestyle='-')
plt.show()
```
