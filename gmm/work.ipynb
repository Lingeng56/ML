```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse
from scipy.stats import multivariate_normal
plt.style.use('seaborn')
```

```python
num1, miu1, var1 = 400, [0.5, 0.5], [1, 3]
x1 = np.random.multivariate_normal(miu1, np.diag(var1), num1)
num2, miu2, var2 = 600, [5.5, 2.5], [2, 2]
x2 = np.random.multivariate_normal(miu2, np.diag(var2), num2)
num3, miu3, var3 = 1000, [1, 7], [6, 2]
x3 = np.random.multivariate_normal(miu3, np.diag(var3), num3)

x = np.vstack((x1, x2, x3))
```

```python
plt.figure(figsize=(10, 8))
plt.axis([-10, 15, -5, 15])
plt.scatter(x1[:, 0], x1[:, 1], s=5)
plt.scatter(x2[:, 0], x2[:, 1], s=5)
plt.scatter(x3[:, 0], x3[:, 1], s=5)
plt.show()


```
